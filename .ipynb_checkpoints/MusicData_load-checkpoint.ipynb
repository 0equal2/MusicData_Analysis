```python
#4. 음악 분석 라이브러리
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
#5. 음악 load
#y,sr=librosa.load(song_path)

#우선 음악 'R_C1'의 10초만 load (duration=10 : 10초 길이)
path='RnBSoul_Calm/R_C1.mp3'
y,sr=librosa.load(path,sr=44100,duration=10)

#음악 load 확인
IPython.display.Audio(data=y, rate=sr)

#음악 plot : x축-time, y축-amplitude

#plt.figure(figsize=(12,3))
#librosa.display.waveplot(y,sr=44100,x_axis='time')
```

```python
path='R_C1_volume_test.mp3'
y,sr=librosa.load(path,sr=44100,duration=10)

#음악 load 확인
IPython.display.Audio(data=y, rate=sr)

#음악 plot : x축-time, y축-amplitude

plt.figure(figsize=(12,3))
librosa.display.waveplot(y,sr=44100,x_axis='time')
```
